```python
from mpl_toolkits.basemap import Basemap  
#安装方法　conda basemap
import numpy as np  
import matplotlib.pyplot as plt  
#创建basemap实例， 参数： 投影模式 经度　纬度　　
m = Basemap(projection='ortho',lat_0=35,lon_0=120,resolution='l')
```

```python
longitudes = [18.96, 24.86, 28.67, 70, 90]  #经度
latitudes = [72.82, 67.01, 77.21, 20, 40]  #纬度
x, y = m(longitudes, latitudes)
```

```python
# 画海岸线，城市边缘，填充大陆
m.drawcoastlines(linewidth=0.25)
m.drawcountries(linewidth=0.25)
# 画投影模式的边缘
m.drawmapboundary(fill_color='#689CD2')
# 每隔30度画经纬线
m.drawmeridians(np.arange(0,360,30))
m.drawparallels(np.arange(-90,90,30))
# 不同的颜色填充大陆
m.fillcontinents(color='#BF9E30',lake_color='#689CD2',zorder=0)
# 绘制散点图，x,y s表示点的大小
m.scatter(x, y, s = 8, marker='o', color='#FF5600')
plt.show()
```
