```python
aoti = mydata.get_data('aotizhongxin_aq','2017-01-01','2018-01-01')
aoti = aoti.fillna(method = 'ffill')
aoti = aoti[['PM2.5',	'PM10',	'NO2','CO',	'O3',	'SO2']]
```

```python
aoti.iloc[1:30]
```

```python
import numpy as np
seq_len = 20
X = []
Y = []
for i in range(len(aoti)-seq_len):
    X.append(np.hstack(np.array(aoti.iloc[i:i+seq_len])))
    Y.append(np.array(aoti.iloc[i+seq_len]))
print(Y)
```

```python
X = np.array(X)
Y = np.array(Y)
```

```python
from sklearn.linear_model import LinearRegression
regr = LinearRegression().fit(X, Y)
regr.predict(X)
```
