```python
from sklearn.linear_model import LinearRegression
from sklearn.externals import joblib
regr = joblib.load("model/train_model.m")
```

```python
ans = {}
```

```python
import pandas as pd
import numpy as np
bj_now = pd.read_csv('data/bj_airquality.csv')
bj_now = bj_now.fillna(method = 'ffill')
bj_now = bj_now.fillna(method = 'bfill')
for i,j in bj_now.groupby(bj_now['station_id']):
    print(i)
    x = np.hstack(np.array(j.iloc[-48:][['PM25','PM10','NO2','CO_Concentration','O3_Concentration', 'SO2_Concentration' ]]))
    y = regr.predict(x.reshape(1,-1))
    ans[i] = y
    #     bj_ans[i] = regr.predict(j.iloc[-48:][['PM25','PM10','NO2','CO_Concentration','O3_Concentration', 'SO2_Concentration' ]])
```

```python
ld_regr = joblib.load("model/ld_train_model.m")
```

```python
ld_now = pd.read_csv('data/ld_aq.csv')
ld_now = ld_now.fillna(method = 'ffill')
ld_now = ld_now.fillna(method = 'bfill')
for i,j in ld_now.groupby(ld_now['station_id']):
    print(i)
    x = np.hstack(np.array(j.iloc[-48:][['PM25','PM10','NO2']]))
    y = ld_regr.predict(x.reshape(1,-1))
    ans[i] = y
```
